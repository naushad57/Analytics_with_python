# EDA

```python
plt.figure(figsize=(10,10))
plt.pie(df['y'].value_counts(),
        labels=df['y'].value_counts());
```

```python
#creating set of subcribers and non-subscriber clients
subscribedc=df[df['y']=='yes'] # subscribed clients
psubscriber=df[df['y']=='no'] #non-subscribed clients
```

```python
plt.figure(figsize=(8,10))
sns.countplot(x=subscribedc['marital']);
```

* married and single are more likely to subscribe in order

```python
plt.figure(figsize=(8,10))
sns.countplot(x=subscribedc['education']);
```

* secondary,tertiary and primary are more likely to subscribe in order

```python
plt.figure(figsize=(10,10))
plt.pie(subscribedc['default'].value_counts(),
        labels=subscribedc['default'].value_counts());
```

* clients with no default are more likely to subscribe

```python
plt.figure(figsize=(10,10))
plt.pie(subscribedc['housing'].value_counts(),
        labels=subscribedc['housing'].value_counts().index,autopct='%1.2f%%');
```

* clients with housing loan are equally likely to subscribe

```python
defaulters=subscribedc[subscribedc['default']=='yes']
cwh=defaulters[defaulters['housing']=='yes'] 
plt.figure(figsize=(10,10))
plt.pie(cwh['housing'].value_counts(),
        labels=cwh['housing'].value_counts());
```

* client is set to default if they have housing loan so avoid contact

```python
plt.figure(figsize=(10,10))
plt.pie(subscribedc['loan'].value_counts(),
        labels=subscribedc['loan'].value_counts().index,autopct='%1.2f%%');
```

* clients with no personal loan are more likely to subscribe

```python
#set of clients who have not defaulted
nzbsubscribedcd=subscribedc[subscribedc['default']=='no']
```

```python
#age of subscribers
plt.figure(figsize=(20,10))
sns.countplot(x=nzbsubscribedcd['age'],order=nzbsubscribedcd['age'].value_counts().index.sort_values());
```

* Clients between 25 - 60 are more likley to subscribe less aged clients being more likely

```python
plt.figure(figsize=(20,10))
sns.countplot(x=nzbsubscribedcd['job']);
```

* clients with management, technician, blue-collar, admin, retired and services are more likely to subscribe in order

```python
#Set of non zero balance subscribed clients
nzbsubscribedc=subscribedc[subscribedc['balance']>0] #eliminating liabilities
```

```python
plt.figure(figsize=(20,10))
plt.plot(nzbsubscribedc['balance']);
```

```python
nzbsubscribedc.describe()
```

* clients with balance more than 1800 are likely to subscribe

```python
plt.figure(figsize=(10,10))
plt.pie(subscribedc['default'].value_counts(),
        labels=subscribedc['default'].value_counts());
```

* client with no default are more likely to subscribe

# list of clients likely to subscriber

```python
pcms=psubscriber[((psubscriber['age']>=25)|(psubscriber['age']<=60))
                 &((psubscriber['job']=='managment')|(psubscriber['job']=='technician'))
                 &((psubscriber['marital']=='married')|(psubscriber['marital']=='single'))
                 &((psubscriber['education']=='secondary')|(psubscriber['education']=='tertiary'))
                 &(psubscriber['default']=='no')
                 &(psubscriber['balance']>1800)
                 &(psubscriber['housing']=='yes')
                &(psubscriber['loan']=='no')]
pcms
```
